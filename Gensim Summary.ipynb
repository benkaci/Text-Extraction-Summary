```python
import gensim 
from gensim.summarization.summarizer import summarize
from gensim.summarization import keywords
from gensim.parsing.preprocessing import remove_stopwords
```

```python
from bs4 import BeautifulSoup 
import requests
```

```python
url="https://www.cancer.net/navigating-cancer-care/cancer-basics/genetics/genetics-cancer"
page= requests.get(url).content

soup = BeautifulSoup(page, 'html.parser')
mydivs = soup.find("div", {"class": "field-items"})

print(mydivs.prettify())
```

```python
paragraphs = soup.find('article').find("div", {'class': 'field-items'}).find_all('p')
for paragraph in paragraphs:
    print (paragraph.text)
    
paragraphs= str(paragraphs)
paragraphs= remove_stopwords(paragraphs)
```

```python
from gensim.summarization.summarizer import summarize
from gensim.summarization import keywords
summ= summarize(paragraphs, ratio=0.3)
print(summ)
```

```python
doc=[]
doc.append(gensim.utils.simple_preprocess(paragraphs, deacc=False, min_len=2, max_len=15))
```

```python
model = gensim.models.Word2Vec(
        doc,
        size=150,
        window=10,
        min_count=1,
        workers=10,
        iter=50
                )
```

```python
model.most_similar("brca") 
```
